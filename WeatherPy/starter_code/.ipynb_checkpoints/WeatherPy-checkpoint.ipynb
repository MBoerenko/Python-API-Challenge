```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
id=[]

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Build URL using the openwweathermap api
base_url = "http://api.openweathermap.org/data/2.5/weather?"

#Set temperature to Farenheit
units = "imperial"
```

```python
#Create the City DataFrame
city_df = pd.DataFrame({})
city_df['City']=""
city_df['Lat']=""
city_df['Lng']=""
city_df['Max Temp(F)']=""
city_df['Humidity(%)']=""
city_df['Cloudiness(%)']=""
city_df['Wind Speed(mph)']=""
city_df['Country']=""
city_df['Date']=""
```

```python
#Set the variables for the for loop
set = 1
limit = 20
set_limit = 0
index=0
```

```python
#Print the heading for the list of cities and numbers
print("Beginning Data Retrieval")
print ("-------------------------------")

#Set up the for loop to loop through the cities
for index, city in enumerate(cities):

    #Create the city dictionary
    city_name = city
    
    #Set up the API to call
    query_url = f"{base_url}&appid={weather_api_key}&units={units}&q="+ city
    
    #Run request
    response = requests.get(query_url)
    
    #Convert to Json
    response_json = response.json()
    
    #Solve for exceptions
    try:
        #Extract the needed fields from the Api
        city_id = response_json['id']
        city_name = response_json['name']
        humidity = response_json['main']['humidity']
        longitude = response_json['coord']['lon']
        latitude = response_json['coord']['lat']
        temperature = response_json['main']['temp_max']
        clouds = response_json['clouds']['all']
        wind_speed = response_json['wind']['speed']
        country = response_json['sys']['country']
        date = response_json['dt']
        
        #Print the records as they are called from the API
        print(f"Processing record {index} of Set {set} |{city_name}/{city_id}")
        
        #Update the sets
        set_limit = set_limit + 1
        
        #Add data to the DataFrame
        city_df.loc[index, 'City'] = city_name
        city_df.loc[index, 'Humidity(%)'] = humidity
        city_df.loc[index, 'Lng'] = longitude
        city_df.loc[index, 'Lat'] = latitude
        city_df.loc[index, 'Max Temp(F)'] = temperature
        city_df.loc[index, 'Cloudiness(%)'] = clouds
        city_df.loc[index, 'Wind Speed(mph)'] = wind_speed
        city_df.loc[index, 'Country'] = country
        city_df.loc[index, 'Date'] = date
         
    except KeyError:
        print("City not found.  Skipping...\n")
    
    #Update the sets
    if set_limit == limit:
        set = set + 1
        set_limit = 0
```

```python
# Create Northern and Southern Hemisphere DataFrames
northern_df = pd.DataFrame({})
southern_df = pd.DataFrame({})
citynotfound_df=pd.DataFrame({})

#create columns in Northern/Southern DataFrames
northern_df['City']=""
northern_df['Humidity(%)']=""
northern_df['Lng']=""
northern_df['Lat']=""
northern_df['Max Temp(F)']=""
northern_df['Cloudiness(%)']=""
northern_df['Wind Speed(mph)']=""

southern_df['City']=""
southern_df['Humidity(%)']=""
southern_df['Lng']=""
southern_df['Lat']=""
southern_df['Max Temp(F)']=""
southern_df['Cloudiness(%)']=""
southern_df['Wind Speed(mph)']=""

citynotfound_df['City']=""
```

```python
#Set up the for loop to loop through the cities
for index, city in enumerate(cities):

    #Create the city dictionary
    city_name = city    
    
    #Set up the API to call
    query_url = f"{base_url}&appid={weather_api_key}&units={units}&q="+ city
    
    #Run request    
    response = requests.get(query_url)
    
    #Convert to Json
    response_json = response.json()

    #Solve for exceptions
    try:
        #Extract the needed fields from the Api 
        city_name = response_json['name']
        humidity = response_json['main']['humidity']
        longitude = response_json['coord']['lon']
        latitude = response_json['coord']['lat']
        temperature = response_json['main']['temp_max']
        clouds = response_json['clouds']['all']
        wind_speed = response_json['wind']['speed']

        #If the latitude is over 0 add to Northern Hemisphere
        if (response_json['coord']['lat']>0): 

            northern_df.loc[index, 'City'] = city_name
            northern_df.loc[index, 'Humidity(%)'] = humidity
            northern_df.loc[index, 'Lng'] = longitude
            northern_df.loc[index, 'Lat'] = latitude
            northern_df.loc[index, 'Max Temp(F)'] = temperature
            northern_df.loc[index, 'Cloudiness(%)'] = clouds
            northern_df.loc[index, 'Wind Speed(mph)'] = wind_speed
            
        #If the latitude is under 0 add to Southern Hemisphere
        else: 

            southern_df.loc[index, 'City'] = city_name
            southern_df.loc[index, 'Humidity(%)'] = humidity
            southern_df.loc[index, 'Lng'] = longitude
            southern_df.loc[index, 'Lat'] = latitude
            southern_df.loc[index, 'Max Temp(F)'] = temperature
            southern_df.loc[index, 'Cloudiness(%)'] = clouds
            southern_df.loc[index, 'Wind Speed(mph)'] = wind_speed
        
    except KeyError:
        citynotfound_df.loc[index, 'City'] = city_name
```
